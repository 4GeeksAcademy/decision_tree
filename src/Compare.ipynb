### 0. Import data and models

```python
import pandas as pd
import os
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
from pickle import dump
import xgboost as xgb
import pickle
```

```python
Xtrain = pd.read_csv('../data/processed/SplitData/NormData/Clean_Xtrain_norm.csv')
Xtest = pd.read_csv('../data/processed/SplitData/NormData/Clean_Xtest_norm.csv')
ytrain = pd.read_csv('../data/processed/SplitData/Clean_ytrain.csv')
ytest = pd.read_csv('../data/processed/SplitData/Clean_ytest.csv')
```

```python
with open("../data/processed/Models/DTree-crit_gini-depth_7-maxf_none-minleaf_4-minsplit_20.sav", "rb") as file:
    DecisionTree_model = pickle.load(file)

with open("../data/processed/Models/RForest-n_100-bootstrap_F-maxf_sqrt.sav", "rb") as file:
    RandomForest_model = pickle.load(file)

Bosting_model = xgb.Booster()
Bosting_model.load_model('../data/processed/Models/Boostrap-n_5-depth_3-g_10.json')
```

```python
Xtrain.columns
```

### 1. Make predictions 
We use the optimal model founded in each option, trained by the same clean data

```python
DT_ypred = DecisionTree_model.predict(Xtest)
DT_accuracy = accuracy_score(ytest, DT_ypred)

RF_ypred = RandomForest_model.predict(Xtest)
RF_accuracy = accuracy_score(ytest, RF_ypred)

B_ypred = Bosting_model.predict(Xtest)
B_accuracy = accuracy_score(ytest, B_ypred)
```

```python
modelos = ['Decision Tree', 'Random Forest', 'Boosting']
accuracies = [DT_accuracy, RF_accuracy, B_accuracy]

plt.figure(figsize=(8, 6))
plt.bar(modelos, accuracies, color=['blue', 'green', 'red'])
plt.xlabel('Modelo')
plt.ylabel('Accuracy')
plt.title('Precisión de los modelos')
plt.ylim(0, 1)
```
